## Problem 1
Fill in the code in the function `make_labels`

```python
def make_labels(data):
    labels = []
    for i in range(len(data)):
        if data[i][0] > 0:
            labels.append(1)
        elif data[i][1] < 0:
            labels.append(-1)
        else:
            labels.append(0)
    return labels
train_labels = make_labels(y_train)
test_labels = make_labels(y_test)
```

## Problem 3
Fill in the code for the profit function to compute the cumulative profit on
a trading strategy based on predictions for the weights in problem 2.

**If predict == 1, then buy[current] and sell[forward]. The profit would be y_train[0].
If predict == 0, no action taken, then no profit.
If predict == -1, then sell[current] and buy[forward]. The profit would be -y_train[1].**

```python
def profit(predictions, data):
    profits = []
    for i in range(len(predictions)):
        if predictions[i] == 1:
            profits.append(data[i][0])
        elif predictions[i] == -1:
            profits.append(-data[i][1])
        else:
            profits.append(0)
    return np.cumsum(profits)
```
